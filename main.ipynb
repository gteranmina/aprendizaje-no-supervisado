# Parte 01: Reducción de Dimensionalidad con Isomap y MDS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import Isomap, MDS

# Ruta local del dataset (actualiza la ruta según la ubicación de tu archivo)
dataset_path = 'dataset(wq).xlsx'
print("Cargando el dataset...")
data = pd.read_excel(dataset_path)
print("Dataset cargado exitosamente.")

# Limpiar datos faltantes (si es necesario)
print("\nLimpiando datos faltantes...")
data = data.dropna()
print("Datos faltantes eliminados.")

# Valores de los parámetros
n_components_list = [3, 5, 7]
neighbors_list = [5, 10]

# Configuración de la figura para los gráficos
print("\nConfigurando la figura para los gráficos...")
fig, axes = plt.subplots(len(n_components_list), len(neighbors_list) + 1, figsize=(20, 15))
fig.suptitle('Reducción de dimensionalidad usando Isomap y MDS', fontsize=16)
print("\nFigura configurada.")

colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown']

# Implementación y visualización de Isomap
for i, n_components in enumerate(n_components_list):
    for j, n_neighbors in enumerate(neighbors_list):
        print(f"\nImplementando Isomap con n_components={n_components}, n_neighbors={n_neighbors}...")
        isomap = Isomap(n_components=n_components, n_neighbors=n_neighbors)
        isomap_data = isomap.fit_transform(data)
        print(f"Isomap con n_components={n_components}, n_neighbors={n_neighbors} completado.")

        ax = axes[i, j]
        ax.scatter(isomap_data[:, 0], isomap_data[:, 1], color=colors[j % len(colors)], alpha=0.7, label=f'n_neighbors={n_neighbors}')
        ax.set_title(f'Isomap (n_components={n_components}, n_neighbors={n_neighbors})')
        ax.legend()

# Implementación y visualización de MDS
for i, n_components in enumerate(n_components_list):
    print(f"\nImplementando MDS con n_components={n_components}...")
    mds = MDS(n_components=n_components, random_state=42)
    mds_data = mds.fit_transform(data)
    print(f"MDS con n_components={n_components} completado.")

    ax = axes[i, len(neighbors_list)]
    ax.scatter(mds_data[:, 0], mds_data[:, 1], color=colors[i % len(colors)], alpha=0.7, label=f'n_components={n_components}')
    ax.set_title(f'MDS (n_components={n_components})')
    ax.legend()

print("\nAjustando y mostrando la figura final...\n")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
print("\nProceso de reducción completado.")

```

# Parte 02: Aplicación de Algoritmos de Clusterización y Evaluación del Desempeño

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import seaborn as sns

print("Iniciando análisis de clustering...")

# Almacenar los resultados de la reducción de dimensionalidad previa
reduced_spaces = {
    'Original': data.values  # Convertir a numpy array
}

# Guardar los resultados de Isomap
for n_comp in n_components_list:
    for n_neigh in neighbors_list:
        isomap = Isomap(n_components=n_comp, n_neighbors=n_neigh)
        isomap_data = isomap.fit_transform(data)
        reduced_spaces[f'Isomap_comp{n_comp}_neigh{n_neigh}'] = isomap_data

# Guardar los resultados de MDS
for n_comp in n_components_list:
    mds = MDS(n_components=n_comp, random_state=42)
    mds_data = mds.fit_transform(data)
    reduced_spaces[f'MDS_comp{n_comp}'] = mds_data

# Crear un DataFrame para almacenar las métricas
metrics_results = []

# Análisis de clustering para cada espacio y cada k
for space_name, space_data in reduced_spaces.items():
    print(f"\nAnalizando clustering para espacio: {space_name}")

    # Probar diferentes valores de k
    for k in range(2, 9):
        print(f"Evaluando k={k}")

        # Aplicar K-means
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(space_data)

        try:
            # Calcular métricas
            silhouette = silhouette_score(space_data, labels)
            davies_bouldin = davies_bouldin_score(space_data, labels)

            # Guardar resultados
            metrics_results.append({
                'Espacio': space_name,
                'K': k,
                'Silhouette Score': silhouette,
                'Davies-Bouldin Index': davies_bouldin
            })
        except Exception as e:
            print(f"Error al calcular métricas para {space_name} con k={k}: {e}")
            continue

# Convertir resultados a DataFrame
metrics_df = pd.DataFrame(metrics_results)

# Visualización de resultados
print("\nCreando visualizaciones...")

# 1. Gráfico de métricas por espacio y k
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))

# Silhouette Score
for space in reduced_spaces.keys():
    space_metrics = metrics_df[metrics_df['Espacio'] == space]
    ax1.plot(space_metrics['K'], space_metrics['Silhouette Score'],
             marker='o', label=space)

ax1.set_title('Silhouette Score por número de clusters')
ax1.set_xlabel('Número de clusters (k)')
ax1.set_ylabel('Silhouette Score')
ax1.grid(True)
ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Davies-Bouldin Index
for space in reduced_spaces.keys():
    space_metrics = metrics_df[metrics_df['Espacio'] == space]
    ax2.plot(space_metrics['K'], space_metrics['Davies-Bouldin Index'],
             marker='o', label=space)

ax2.set_title('Davies-Bouldin Index por número de clusters')
ax2.set_xlabel('Número de clusters (k)')
ax2.set_ylabel('Davies-Bouldin Index')
ax2.grid(True)
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

# 2. Visualización de clusters para el mejor k en cada espacio
# Encontrar el mejor k basado en Silhouette Score para cada espacio
best_k = {}
for space in reduced_spaces.keys():
    space_metrics = metrics_df[metrics_df['Espacio'] == space]
    if not space_metrics.empty:
        best_k[space] = int(space_metrics.loc[space_metrics['Silhouette Score'].idxmax(), 'K'])

# Crear visualización de clusters
print("\nVisualizando clusters para los mejores valores de k...")
n_spaces = len(reduced_spaces)
n_rows = (n_spaces + 2) // 3  # Ajustar el número de filas según la cantidad de espacios
fig, axes = plt.subplots(n_rows, 3, figsize=(20, 5*n_rows))
if n_rows == 1:
    axes = axes.reshape(1, -1)

# Aplanar el array de axes para facilitar la iteración
axes_flat = axes.flatten()

for idx, (space_name, space_data) in enumerate(reduced_spaces.items()):
    if idx < len(axes_flat) and space_name in best_k:
        k = best_k[space_name]
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(space_data)

        # Verificar que tenemos al menos 2 dimensiones para visualizar
        if space_data.shape[1] >= 2:
            scatter = axes_flat[idx].scatter(space_data[:, 0], space_data[:, 1],
                                           c=labels, cmap='viridis')
            axes_flat[idx].set_title(f'{space_name}\nMejor k={k}')

            # Añadir centroides
            centers = kmeans.cluster_centers_
            axes_flat[idx].scatter(centers[:, 0], centers[:, 1],
                                 c='red', marker='x', s=200, linewidth=3,
                                 label='Centroids')
            axes_flat[idx].legend()

# Ocultar los subplots vacíos
for idx in range(len(reduced_spaces), len(axes_flat)):
    axes_flat[idx].set_visible(False)

plt.tight_layout()
plt.show()

# Imprimir resultados detallados
print("\nResultados detallados de las métricas:")
print(metrics_df.to_string(index=False))

print("\nMejores valores de k por espacio (basado en Silhouette Score):")
for space, k in best_k.items():
    best_metrics = metrics_df[(metrics_df['Espacio'] == space) & (metrics_df['K'] == k)]
    print(f"\nEspacio: {space}")
    print(f"Mejor k: {k}")
    print(f"Silhouette Score: {best_metrics['Silhouette Score'].values[0]:.4f}")
    print(f"Davies-Bouldin Index: {best_metrics['Davies-Bouldin Index'].values[0]:.4f}")
```
